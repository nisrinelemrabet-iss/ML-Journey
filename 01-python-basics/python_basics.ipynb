```python
grades = [15, 16, 14, 17, 18]

print("Grades:", grades)
print("First grade:", grades[0])
print("Average:", sum(grades) / len(grades))
```

```python
for grade in grades:
    if grade >= 16:
        print(grade, "→ Excellent")
    else:
        print(grade, "→ Needs improvement")
```

FUNCTIONS :

```python
def average(numbers):
    return sum(numbers) / len(numbers)

print("Average grade:", average(grades))
```
